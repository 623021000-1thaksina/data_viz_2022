```python
import pandas as pd #จัดการตารางต่างๆ
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv') 
df
```

```python
df['Name'].unique() #unique เป็นการเลือกมาเฉพาะค่าที่ไม่ซ้ำ
```

iris-dataset.png

# Parallel Coordinates

```python
pd.plotting.parallel_coordinates(df, 'Name')  # (input,ชื่อคอลัมน์) #pandas.plotting.parallel_coordinates(fame, class_colum) 
```

### select columns to show

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']) #cols เลือกดูเฉพาะคอลัมที่จะนำมาใช้พล็อต
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']); # ; ใช้เพื่อลบตัวที่จะปริ้นเป็นข้อความทั้งหมด (โชว์แต่รูป)
```

### change color

https://matplotlib.org/stable/tutorials/colors/colors.html

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['r','y','k'])
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]])
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['#FF00E0', '#00FFE2', '#ECFF00'])
```

### Specify color by colormap
https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
pd.plotting.parallel_coordinates(df, 'Name', colormap='Set1')
```

## Subplot

```python
matplotlib.rcParams['figure.figsize']=[10,10]
```

```python
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,3)
plt.imshow(df.iloc[100:150,:-1],interpolation='nearest',cmap='hot')
```

```python
plt.subplot(1,4,1)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,4,2)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.2)

plt.subplot(1,4,3)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.5)

plt.subplot(1,4,4)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.8)
```
